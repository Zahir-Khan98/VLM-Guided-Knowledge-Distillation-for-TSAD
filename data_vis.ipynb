```python
import matplotlib.pyplot as plt
import numpy as np
data=np.load('pems04.npz','rb')

print(data.files)
array_one=data['data']
print(array_one.shape)
```

# Dataset visualization for 10 channels
## 1. stacked view 

```python
plt.figure(figsize=(10,20))
for i in range(10):
    
    plt.subplot(10,1,i+1)
    plt.plot(array_one[:100,i,0])

plt.show()
```

## 2. Overlay view

```python
plt.figure(figsize=(10,15))
for i in range(10):
    plt.plot(array_one[:100,i,0])
```
